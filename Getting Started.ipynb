# Saving your work with Git (and GitHub)

Git is a version control program that is in every Linux machine by default. To use it, you must initialize a directory to create a new *repository* using the command:
```
git init
```

To add a file to your list of tracked files, you use the following command:
``` 
git add <name of file or directory you want to track>
```

To save the current version of your tracked files, you use the following command:
``` 
git commit -am 'add a quick comment about the changes you have made'
```

Once you have created a new commit, you can synchronize your local repository with your GitHub repository on the cloud. You can do this using *git push*. The local repo is called 'origin' by default, and the remote repository on GitHub is called 'master' by default. The command to send your commit to the cloud looks like this:
``` 
git push origin master
```
