# Removing adapters using cutadapt

## Riboseq data


**Reads structure**:     RRRR(NNNN...NNNN)RRRR-adaptor-

**R** represents random nucleotides. Random nucleotides serve as unique molecular identifiers (UMIs), **N** corresponds to the sequence of the RNA protected fragment, and the **adaptor** sequence is TGGAATTCTCGGGTGCCAAGG

### First step: Remove initial N and adaptor

```python
%%bash
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/fastq_files/"
extention=".fastq"
addition="_1"
label="_N_adaptor"
directory="/projects_eg/projects/simone/Duncan_Mata/trimming/"
mkdir $directory
output_directory="/projects_eg/projects/simone/Duncan_Mata/trimming/Ribo_N_adaptor/"
mkdir $output_directory



list='61 62 65 66 69 70 73 74'
accession_base="ERR19949"
for i in $list
do 
cutadapt --trim-n -b file:/home/simone/Downloads/Trimmomatic-0.36/adapters/Duncan_Ribo_adaptor.fa -o $output_directory$accession_base$i$label$extention $path_fastq_files$accession_base$i$addition$extention
done


# (-O) Minimum number of nucleotides of the adaptor to trim. By default is 3.
```

### Second step: Remove PCR duplicate to generate a non-redundant dataset

```python
%%bash
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/trimming/Ribo_N_adaptor/"
extention=".fastq"    # no extention for output
label="_N_adaptor"
label2="_no_dup"
output_directory="/projects_eg/projects/simone/Duncan_Mata/trimming/Ribo_no_dup/"
mkdir $output_directory





list='61 62 65 66 69 70 73 74'
accession_base="ERR19949"
for i in $list
do 
perl /home/simone/Downloads/prinseq-lite-0.20.4/prinseq-lite.pl -out_good $output_directory$accession_base$i$label2 -fastq $path_fastq_files$accession_base$i$label$extention -derep 1 -no_qual_header
done

# -derep 1  --> remove exact duplicate
```

### Third step: Remove first four and last four random nucleotides

```python
%%bash
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/trimming/Ribo_no_dup/"
extention=".fastq"
label="_no_dup"
label2="_trimmed"
output_directory="/projects_eg/projects/simone/Duncan_Mata/trimming/Ribo_trimmed/"
mkdir $output_directory


# remove first four and last four nucleotides
list='61 62 65 66 69 70 73 74'
accession_base="ERR19949"
for i in $list
do 
cutadapt -u 4 -u -4 -m 26 -o $output_directory$accession_base$i$label2$extention $path_fastq_files$accession_base$i$label$extention
done

```

#### FastQC

```python
%%bash 
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/trimming/Ribo_trimmed/"
extention=".fastq"
label="_trimmed"
output_directory="/projects_eg/projects/simone/Duncan_Mata/fastqc_reports/Ribo_final/"
mkdir $output_directory


list='61 62 65 66 69 70 73 74'
accession_base="ERR19949"
for i in $list
do 
fastqc -I -O $output_directory $path_fastq_files$accession_base$i$label$extention
done

cd /projects_eg/projects/simone/Duncan_Mata/fastqc_reports/Ribo_final/
multiqc .
```

## RNAseq data

**Reads structure**:     RRRRRRRRR(NNNN....NNNN)

**R(9)** corresponds to a UMI and **N** to the sequence of fragmented RNA molecules

```python
%%bash
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/fastq_files/"
extention=".fastq"
addition="_1"
label="_adaptor"
output_directory="/projects_eg/projects/simone/Duncan_Mata/trimming/RNA_adaptor/"
mkdir $output_directory



list='59 60 63 64 67 68 71 72'
accession_base="ERR19949"
for i in $list
do 
cutadapt -b file:/home/simone/Downloads/Trimmomatic-0.36/adapters/Duncan_RNA_adaptors.fa -o $output_directory$accession_base$i$label$extention $path_fastq_files$accession_base$i$addition$extention
done
```

### Second step: Remove PCR duplicate to generate a non-redundant dataset

```python
%%bash
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/trimming/RNA_adaptor/"
extention=".fastq"  # no extention for output
label="_adaptor"
label2="_no_dup"
output_directory="/projects_eg/projects/simone/Duncan_Mata/trimming/RNA_no_dup/"
mkdir $output_directory





list='59 60 63 64 67 68 71 72'
accession_base="ERR19949"
for i in $list
do 
perl /home/simone/Downloads/prinseq-lite-0.20.4/prinseq-lite.pl -out_good $output_directory$accession_base$i$label2 -fastq $path_fastq_files$accession_base$i$label$extention -derep 1 -no_qual_header
done
```

### Third step: Remove first nine random nucleotides

```python
%%bash
path_fastq_files="/projects_eg/projects/simone/Duncan_Mata/trimming/RNA_no_dup/"
extention=".fastq"
label="_no_dup"
label2="_trimmed"
output_directory="/projects_eg/projects/simone/Duncan_Mata/trimming/RNA_trimmed/"
mkdir $output_directory


# remove first nine nucleotides
list='59 60 63 64 67 68 71 72'
accession_base="ERR19949"
for i in $list
do 
cutadapt -u 9 -q 30 -m 26 -o $output_directory$accession_base$i$label2$extention $path_fastq_files$accession_base$i$label$extention
done
```
